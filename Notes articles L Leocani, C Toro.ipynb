# Article : Cohérence lie à event desync /syn liée à EEG, L. Leocani , C.Toro

**Analyse EEG** :
1. Données inutilisables : celles qui comportent artefacts, exécution des mouvements pas assez rapide, relaxation musculaire incomplète entre 2 mouvements.
2. Découpage du signal EEG : découpage en époque de 500 ms qui permet une résolution fréquentielle de 2 Hz
3. Fenêtres glissantes : avec un recouvrement de 250ms courant la période allant de - 2,55 à + 2,5s
4. Nombre époque identiques : comme artefact sont éliminés, le nombre d'époques restent identiques pour chaque intervalles.

cohérence dépend du nombre d'époques utilisées car plus on a d'échantillon plus ça diminue la valeur de la cohérence → faire tu, moyenne

5. Conditions de références : spectre de puissance FFT obtenu durant époque la plus précoce
obtenu durant époque la plus précoce
6. Sélectionner des bandes de fréquences : pour étude temporelle de la puise spectre, cohérence et cohérence de phase 
Formule puissance :
((X-baseline)/baseline) x 100

**Analyse de la cohérence**
calculée pour chaque paires d'électrodes à partir des spectres de puissance obtenus par Fourrier Transform (FFT). 
cohérence = corrélation croisée de 2 spectres de puissance divisé par les puissances respectives → méthode qui évite les biais

**Analyse de la cohérence de phase en millisecondes**
1. Calcul des moyennes de groupe pour les variations relatives de puissance, différences de cohérence, les mesures de cohérence de phases
Les aires :
• aire sensori-motrice contralaterale :
motricité + sensibilité du côté oppose
• aire frontale bilatérale :
avant du cerveau + les deux côte du cerveau en même temps
• région centrale / temporales contralateral
zone centrale + temporale de hémisphère opposé du corps concerne
• region centro - temporo-parietales:
zone centrale + temporale + pariétale

synchronisation plutôt dans la bande 18-22hz

**épreuves d'activation EEG** :
hyperpnée = respiration ample répétée de façon continue durant 5 minutes
SLI (simulation lumineuse intermittente)= flash lumineux, 30 cm yeux patient , à différentes fréquences 
enregistrement EEG de sieste : le but de favoriser les enregistrements d'anomalie épileptique 
qEEG : quantification du signal par opposition à l'analyse qualitative par lecture visuelle

Amplitude : quantification la plus simple, calcul de la max, min, moyenne.
Puissance spectrale : extraire du contenu d'un signal, permet visualisation tridimensionnel
Complexité : quantification de la complexité du signal
Connectivité : quantification du partage d’information entre 2 électrodes
