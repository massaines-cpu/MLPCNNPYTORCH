# CNN
- architecture spécifique pour apprentissage automatique et deep learning
- dispose de plusieurs couches comme la **couche de convolution**
- permet de comprendre les structures visuelles complexes, classification d'images

# Couches de convolution

objectif : extraire des caractéristiques significatives de données, reconnaitre motifs ou pattern

filtres sont appliqués sur les données d'entrée

# Couche de pooling
une image est une matrice de nombres, chaque nombre représente l'intensité d'un pixel
objectif : réduire la dimension spatiale, réduire le nombre de param du modèle, créer invariance aux petites translations (translation = déplacement de l'objet dans l'image donc CNN reconnait l'objet même s'il a bougé) préserver les valeurs dominantes

2 types de pooling :
1. pooling max = retient la valeur max dans chaque région
2. pooling moyen = calcule la moyenne des valeurs

# Couches Fully-Connected (FC)

objectifs : permettent une connexion complète entre ensemble des unités neuronales de la couche précédente

# Fonctionnement

CNN reçoit un input (image sous forme de matrice de pixels) -> applications de filtre sur l'image (s'applique à des région local, coin, texture, motif) -> fonction d'activation (pour introduire non linéarité dans modèle), ReLU --> couche pooling pour réduire complexité du modèle --> étape des couches se répète --> caractéristique extraites passe dans couches connectées -> prédiction ou classification
