# Les types de fonction d'activation

- fonction sigmoïde est idéale pour les problèmes de classification binaire f(x) = 1 / (1 + e^-x), limite = problème du "gradient disparaissant"
- fonction softmax est utile pour les prédictions multi-classes
- fonction ReLU permet de surmonter le problème du gradient qui s'évanouit
- fonction tanh f(x) = (e^x - e^-x) / (e^x + e^-x) traite les valeurs négative plus efficacement que la sgmoide et est centré sur zéro, valeurs symétriques, même limite que sigmoide
